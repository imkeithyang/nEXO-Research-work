```python
from mpl_toolkits import mplot3d

%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt


x = []
y = []
z = []

x.append([])
y.append([])
z.append([])
idx = 0
with open("/Users/Tiger/nexo_py/txt/bb0n_x.txt") as f1, open("/Users/Tiger/nexo_py/txt/bb0n_y.txt") as f2, open("/Users/Tiger/nexo_py/txt/bb0n_z.txt") as f3:
    for (line1,line2,line3) in zip(f1,f2,f3):
        if( line1 != "seprate\n"):
            x[idx].append( float(line1[:len(line1)-1]) )
            y[idx].append( float(line2[:len(line1)-1]) )
            z[idx].append( float(line3[:len(line1)-1]) )
        else:
            idx+=1
            #if(idx == 10000):
                #break
            x.append([])
            y.append([])
            z.append([])
    
```
